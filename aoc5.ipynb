```python
with open("5.txt", "r") as f:
    ranges_raw, ids_raw = f.read().strip().split("\n\n")
```

```python
ids = ids_raw.strip().splitlines() |> map[int] |> set |> sorted
```

```python
# part 1
ranges = [0]
for start, end in ranges_raw.strip().splitlines() |> map[$.strip().split("-")] |> map[map[int] .> list] |> sorted:
    if start <= ranges[-1]:
        ranges[-1] = max(ranges[-1], end)
    else:
        ranges.extend([start-1, end])
ranges.pop(0)

ids |> map[np.searchsorted(ranges, $)] |> map[$ % 2 == 1] |> np.sum
```

```python
# part 2
zip(ranges[1::2], ranges[::2]) |> map[$ *|> $ - $] |> sum
```

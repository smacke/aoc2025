```python
# part 2
ans = 0
while True:
    padded = np.array(orig)
    for i in (-1, 0, 1):
        for j in (-1, 0, 1):
            if i==0 and j==0:
                continue
            padded += np.roll(orig, shift=(i, j), axis=(0, 1))
    removable = (padded <= 4) * orig
    num_removable = np.sum(removable)
    if num_removable == 0:
        break
    ans += num_removable
    orig -= removable
```

```python
ranges = inp.strip().split(",") |> map[$.strip().split("-") *|> [int($)-1, int($)]] |> sorted |> sum($, []) |> np.array
```

```python
ans = 0
for v in itertools.count(1):
    vv = int(2*str(v))
    idx = np.searchsorted(ranges, vv)
    if idx == len(ranges):
        break
    elif idx % 2 == 1:
        ans += vv
ans
```

```python
# part 2
invalid = set()
finished = False
for v in itertools.count(1):
    for j in itertools.count(2):
        vv = int(j*str(v))
        idx = np.searchsorted(ranges, vv)
        if idx == len(ranges):
            finished = j == 2
            break
        elif idx % 2 == 1 and vv not in invalid:
            invalid.add(vv)
    if finished:
        break
sum(invalid)
```

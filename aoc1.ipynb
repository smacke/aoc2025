```python
debug_inp = """
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
""".strip().splitlines()
```

```python
with open("1.txt", "r") as f:
    full_inp = f.read().strip().splitlines()
```

```python
def part1(inp):
    cur = 50
    ans = 0
    for v in inp:
        sgn = (v[0] == "R")*2 - 1
        cur = (cur + sgn * int(v[1:])) % 100
        ans += cur == 0
    return ans
```

```python
def part2(inp):
    cur = 50
    ans = 0
    for v in inp:
        sgn = (v[0] == "R")*2 - 1
        dist = int(v[1:])
        ans += (dist + (100 + sgn * cur) % 100) // 100
        cur = (cur + sgn * dist) % 100
    return ans
```

```python
debug_inp |> part1
```

```python
full_inp |> part1
```

```python
debug_inp |> part2
```

```python
full_inp |> part2
```

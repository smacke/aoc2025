```python
with open("12.txt", "r") as f:
    inp = f.read().strip().split("\n\n")
```

```python
shapes, rest = inp[:6], inp[6]
```

```python
n_hashes = shapes |> map[$ |> map[$ == "#"] |> sum]
```

```python
with allow_pipelines_in_loops_and_calls():
    valid = 0
    for idx, line in enumerate(rest.strip().splitlines()):
        dims, counts = line.strip().split(": ")
        area = dims.split("x") |> map[int] |> reduce[$ * $]
        counts = counts.strip().split() |> map[int]
        if area >= sum(counts) * 9:
            valid += 1
        elif sum(n_hashes[i] * c for i, c in enumerate(counts)) > area:
            # impossible case
            pass
        else:
            raise ValueError("Unable to handle case at index with simple checks %d" % idx)
valid
```

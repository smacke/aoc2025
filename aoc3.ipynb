```python
import numpy as np
```

```python
inp = """
987654321111111
811111111111119
234234234234278
818181911112111
"""
```

```python
with open("3.txt", "r") as f:
    inp = f.read()
```

```python
banks = inp.strip().splitlines() |> map[$ |> map[int] |> list]
```

```python
# part 1
ans = 0
for bank in banks:
    idx = np.argmax(bank[:-1])
    val = bank[idx] * 10 + max(bank[idx+1:])
    ans += val
ans
```

```python
# part 2
ans = 0
for bank in banks:
    parts = []
    for j in range(12):
        if j == 11:
            idx = np.argmax(bank)
        else:
            idx = np.argmax(bank[:j-11])
        parts.append(bank[idx])
        bank = bank[idx + 1:]
    ans += int("".join(str(p) for p in parts))
ans
```

```python
with open("8.txt", "r") as f:
    inp = f.read().strip().splitlines()
connections = 1000
```

```python
def find(x):
    while parent[x] != x:
        x, parent[x] = parent[x], parent[parent[x]]
    return x

def union(x, y):
    x = find(x)
    y = find(y)
    if x == y:
        return
    if rank[x] < rank[y]:
        x, y = y, x
    parent[y] = x
    rank[x] = max(rank[x], rank[y] + 1)
    size[x] += size[y]
```

```python
coords = inp |> map[$.split(",") |> map[int] |> np.array($, dtype=int)]
```

```python
N = len(inp)
parent = np.arange(N)
size = np.ones(N, dtype=int)
rank = np.zeros(N, dtype=int)

q = []
for i in range(N):
    for j in range(i):
        q.append((np.sum((coords[i] - coords[j])**2), (i, j)))

for idx, (_, (i, j)) in enumerate(sorted(q)):
    were_disconnected = find(i) != find(j)
    union(i, j)
    N -= were_disconnected
    if N == 1 and were_disconnected:
        print("part 2:", coords[i][0] * coords[j][0])
    if idx == connections - 1:
        seen = set()
        circuits = []
        for i in range(N):
            root = find(i)
            if root in seen:
                continue
            seen.add(root)
            circuits.append(size[root])
    if N == 1 and idx > connections:
        break
```

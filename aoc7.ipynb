```python
inp = """
.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
""".strip().splitlines()
```

```python
with open("7.txt", "r") as f:
    inp = f.read().strip().splitlines()
```

```python
inp = inp |> map[$ |> list |> ["."] + $ + ["."]]
```

```python
paths = [0] * len(inp[0])
num_paths = 1
num_splits = 0
paths[inp[0].index("S")] = 1
for i in range(1, len(inp)):
    newpaths = [0] * len(paths)
    for j in range(len(inp[i])):
        if not paths[j]:
            continue
        if inp[i][j] == "^":
            newpaths[j-1] += paths[j]
            newpaths[j+1] += paths[j]
            num_paths += paths[j]
            num_splits += 1
        else:
            newpaths[j] += paths[j]
    paths = newpaths
print(num_splits, num_paths)
```
